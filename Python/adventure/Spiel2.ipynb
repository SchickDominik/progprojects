````python
import random
from time import sleep

class Stack(): #Gegner-Management
    def __init__(self):
        self.s = []
    def push(self, x):
        self.s.append(x)
    def pop(self):
        self.s.pop()
    def pop_front(self):
        x = self.s[0]
        self.s = self.s[1:]
        return x
    def randomize(self):
        x = random.shuffle(self.s)
        self.s = x
    def to_list(self):
        return self.s

class Player():
    def __init__(self):
        self.health_max = 5
        self.health_cur = self.health_max
        self.gold = 30 #Startgeld (hat noch keinen Sinn -> Future-Feature)
        self.weapon = Gegenstand(1)
        self.level = 1 #Hat zZ keinen Einfluss (Future-Feature)
        self.traenke = 1
        self.xp = 0
        # self.reset()

    def lvl_up(self):
        self.level += 1
        print('\033[92m'+"Du bist auf Level "+str(self.level)+" aufgestiegen!"+'\033[0m')
        self.health_max +=30
        self.health_cur +=30
        self.xp -= 100

    def health(self):
        print("Deine Gesundheit: "+str(self.health_cur)+"/"+str(self.health_max))

    def reset(self):
        self.__init__()
        # self.health_max = 20
        # self.health_cur = self.health_max
        # self.gold = 30 #Startgeld (hat noch keinen Sinn -> Future-Feature)
        # self.weapon = Gegenstand(1)
        # self.level = 1 #Hat zZ keinen Einfluss (Future-Feature)
        # self.traenke = 1
        # self.xp = 0

class NPC():
    def __init__(self,health,strength):
        self.health = health*random.randrange(10,21) 
        self.strength = strength*random.randrange(5,11)
    def health_print(self):
        print("Gegner Gesundheit: "+str(self.health))
        
class Bandit(NPC):
    def __init__(self):
        super().__init__(1,1) #set health + strength
        self.name = "Bandit"
        self.quip = ["Hey!", "AUA!", "Aaargh!", "Autsch!"]
        self.accuracy = 0.65
        self.loot_chance = 0.80
        self.loot_level = 2
        self.xp_drop = random.randrange(30,41)

class Ork(NPC):
    def __init__(self):
        super().__init__(3,3)
        self.name = "Ork"
        self.quip = ["Uagh!", "Uff!", "Gnar!", "Grrrrr!"]
        self.accuracy = 0.75
        self.loot_chance = 40
        self.loot_level = 3
        self.xp_drop = random.randrange(50,71)

class Riese(NPC):
    def __init__(self):
        super().__init__(10,10)
        self.name = "Riese"
        self.quip = ["AAAAAh!", "Fi, Fai, Fo, Fumm!", "Hargh!"]
        self.accuracy = 0.9
        self.loot_chance = 30
        self.loot_level = 5
        self.xp_drop = random.randrange(100,151)
        
class Gegenstand():
    def __init__(self, level):
        adjective = ["Schmutzige", "Krumme", "Große", "Alte", "Glanzvolle", "Blutige"]
        obj_type = ["r Degen", "r Hammer", "r Dolch", "r Bogen", "r Speer", "s Schwert", "r Ast", " Axt"]
        self.level = level
        self.name = random.choice(adjective) + random.choice(obj_type)
        self.damage = random.randrange(3,6)*level #=> 3-5
    
    def __repr__(self):
        return "Deine Waffe: "+self.name+", mit "+str(self.damage)+" Schaden"
    
class Level():
    def __init__(self,orks,bandits,riesen):
        self.orks = orks
        self.bandits = bandits
        self.riesen = riesen
        self.enemies=[]
        for i in range(self.orks):
            self.enemies.append(Ork())
        for i in range(self.bandits):
            self.enemies.append(Bandit())
        random.shuffle(self.enemies)
        for i in range(self.riesen):
            self.enemies.append(Riese())

        
class Level_1(Level):
    def __init__(self):
        super().__init__(0,4,0) #spawnt 0 Orks und 4 Banditen
        #self.enemies.push(boss) #Future-Feature
        self.welcome = "Willkommen zu deinem Abenteuer! Erkundungslustig machst du dich auf die Welt zu erkunden. Du hast einen Trank dabei. Dein Vater gab dir außerdem etwas mit, das dich schützen soll. "+str(spieler.weapon)+". Du kommst an einen dunklen Wald."
        self.story_bit = ["Ein Schatten schleicht sich an...", "Du hörst Geräusche aus dem Unterholz.", "Du spürst Blicke in deinem Rücken", "Und plötzlich...!!"]
        self.goodbye = "Puh, geschafft! Du entfernst dich vom dunklen Wald und machst dich bereit für dein nächstes Abenteuer."

class Level_2(Level):
    def __init__(self):
        super().__init__(2,3,0)
        self.welcome = "Du kommst aus dem Wald heraus in eine bergige Region. Du erkennst einen kleinen Ort in der Ferne. Kampferprobt machst du dich auf."
        self.story_bit = ["Hinter einem Stein lauert etwas...", "Wie aus dem Nichts..!", "Schritte...", "Ein Hinterhalt!", "Doch was ist das für eine Gestalt?"]
        self.goodbye = "Völlig außer Atem erreichts du das kleine Bergdorf. Du ruhst dich in der Taverne erst mal aus."
        
class Level_3(Level):
    def __init__(self):
        super().__init__(4,0,0)
        self.welcome = "Der Wirt erzählt dir von einem Ork-Lager. Sie hätten einen großen Schatz gestohlen. Du fühlst dich heldenhaft und bietest deine Hilfe an. Auf geht's!"
        self.story_bit = ["Eine Lagerwache sieht dich!", "Aus einer Hütte springt dich etwas an!", "Du hörst Brüllen!", "Schnappt ihn euch!!"]
        self.goodbye = "Du hast die Orks geschlagen. Du suchst das Lager ab und findest eine gewaltige Truhe. In dieser liegt eine Schriftrolle: Freunde sind der wahre Schatz!"
        
class Level_4(Level):
    def __init__(self):
        super().__init__(3,0,1)
        self.welcome = "Enttäuscht und wütend suchst du den Wirt auf. Er versichert dir, es war ein Versehen. Dieses Mal aber wartet der große Schatz auf dich. Auf der Spitze des Berges wohnt ein Riese. Warum nicht? Du machst dich auf und besteigst den Berg."
        self.story_bit = ["Aus einer Höhle schleicht sich etwas heraus.", "Ein Stein trifft dich am Hinterkopf!", "Was in aller Welt..."]
        self.goodbye = "Der Riese ist erlegt. Du suchst den vermeintlichen Schatz. Im Augenwinkel siehst du noch, wie der Wirt hinter einem Stein hervor springt und dich niederschlägt. Er stiehlt den Schatz... und sich davon."

spieler = Player()
def game_loop():
    #score = 0 #Future-Feature
    game_over = False
    spieler.reset()
    level=[Level_1(),Level_2(),Level_3(),Level_4()] #Alle Level als Klassen hier rein
    while not game_over:
        i=1
        for welt in level:
            if encounter(welt,i) == 0:
                i+=1
            else:
                break
        game_over = True

    print("Das Spiel ist vorbei!")
    #print(score)
    retry = input("Noch ne Runde? j/n")
    if retry.lower() == "j":
        game_loop() #Rekursion zum erneuten Start
    
    
def encounter(welt,i):
    dead = 0
    aktuelles_level = welt
    print(" ")
    print("*******************")
    print('\033[1m'+"Kapitel "+str(i)+'\033[0m')
    print("")
    print(aktuelles_level.welcome)
    for enemy in aktuelles_level.enemies:
        print("_____________________")
        print(random.choice(aktuelles_level.story_bit))
        if fight(enemy) == 0:
            options()
        else:
            dead = 1
            break
    if dead == 0:
        print(aktuelles_level.goodbye)
    return dead

def fight(Gegner):
    flee = 0
    dead = 0
    while dead == 0:
        while Gegner.health > 0 and flee == 0:
            print("                 ")
            print("Du kämpfst gegen: "+Gegner.name)
            Gegner.health_print()
            spieler.health()
            a = fight_options()
            if a == 1: #Angriff #Genauigkeit?
                print("Du greifst mit deiner Waffe an und "+'\033[94m'+"machst "+str(spieler.weapon.damage)+" Schaden."+'\033[0m')
                Gegner.health -= spieler.weapon.damage
                print("Gegner: "+random.choice(Gegner.quip))
                if Gegner.health <= 0:
                    print("       ")
                    print('\033[92m'+"**** Du hast gewonnen! ****"+'\033[0m')
                    print(" ")
                    break
            elif a == 2: #Trank
                if spieler.traenke > 0:
                    if spieler.health_max - spieler.health_cur < 30:
                        spieler.health_cur = spieler.health_max
                        print("Du nimmst einen Trank ein.")
                    else:
                        spieler.health_cur += 30
                    spieler.traenke -=1
                else:
                    print("Du hast keine Tränke.")
            elif a == 3: #Fluchtversuch
                f = random.random()
                if f < 0.6: # 60% Fluchchance (fix? Future-Feature)
                    flee = 1
                    break
                else:
                    print("Dein Gegner ist schnell... Der Kampf geht weiter!")
            else:
                print("Bitte gib 1, 2, oder 3 ein um fortzufahren.")
            g_a = random.random()
            if g_a < Gegner.accuracy:
                print("Dein Gegner greift an und "+'\033[91m'+"macht "+str(Gegner.strength)+" Schaden."+'\033[0m')
                spieler.health_cur -= Gegner.strength
                if spieler.health_cur <= 0:
                    print('\033[91m'+"Du wurdest besiegt, dein Abenteuer ist vorbei!"+'\033[0m')
                    print(" ")
                    game_over = True
                    dead = 1
                    break
            else:
                print("Dein Gegner greift an.")
                #sleep(1)
                print("Verfehlt")

        if flee == 1 and dead == 0:
            print('\033[92m'+"Du bist entkommen. Du erhältst keine Belohnung."+'\033[0m')
            break
        elif dead == 1:
            break
        else:
            l = random.random()
            if l < Gegner.loot_chance: #Chance ob Loot-Drop oder nicht (abhängig von Gegner)
                loot(Gegner)
                gold_reward = Gegner.loot_level * random.randrange(10,20)
                spieler.gold += gold_reward
                print("Du erhältst "+'\033[93m'+str(gold_reward)+" Gold."+'\033[0m')
                spieler.xp += Gegner.xp_drop
                print("Du erhältst "+str(Gegner.xp_drop)+" Erfahrungspunkte.")
                if spieler.xp > 100:
                    spieler.lvl_up()
            else:
                print("Scheint als wärst du leer ausgegangen.")
        break 
    return dead 

def loot(Gegner):
    l = random.random() #Chance ob Gegenstand oder Trank (60/40)
    drop = ["in den Graben.", "in die Büsche.", "auf einen Baum.", "in den Fluss.", "in das nächste Toilettenhäuschen."]
    if l < 0.60:
        g = Gegenstand(Gegner.loot_level)
        print("Du findest: "+g.name)
        print("Schaden: "+str(g.damage))
        print("Möchtest du die Waffe eintauschen? j/n")
        ant=input()
        print(" ")
        if ant.lower() == "j":
            spieler.weapon = g
            print("Du wirfst deine alte Waffe "+random.choice(drop))
    else:
        print("Du erhältst einen Trank.")
        spieler.traenke += 1

def options():
    print("        ")
    print("Du machst kurz Pause.")
    print("Wähle aus: 1/2/3")
    print("1. Inventar")
    print("2. Tränke")
    print("3. Weiter")
    x = input()
    opt = int(x)
    print(" ")
    if opt == 1:
        print(spieler.weapon)
        print("  ")
        options()
    elif opt == 2:
        if spieler.traenke > 0:
            print("Deine Tränke: "+str(spieler.traenke))
            spieler.health()
            print("Möchtest du einen Trank nehmen? j/n")
            ant = input()
            if ant.lower() == "j":
                if spieler.health_cur == spieler.health_max:
                    print("Du bist bei voller Gesundheit.")
                else:
                    if spieler.health_max - spieler.health_cur < 30:
                        spieler.health_cur = spieler.health_max
                    else:
                        spieler.health_cur += 30
                    spieler.traenke -= 1
            options()
            
        else:
            print("Du hast keine Tränke.")
            options()
    elif opt !=3:
        print("Wähle aus: 1/2/3")

def fight_options():
    print("Wähle aus: 1/2/3")
    print("1. Angriff mit Waffe")
    print("2. Trank nehmen")
    print("3. Fluchtversuch")
    ant = input()
    return int(ant)
````
